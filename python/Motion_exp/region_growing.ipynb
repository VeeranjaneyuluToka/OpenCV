```python
from collections import deque
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
video_file_path = '../../../../installation/exp/OF/data/checkerball.mp4'
redlower = (6, 10, 90)
redupper = (20, 20, 255)
```

```python
cap = cv.VideoCapture(video_file_path)

if not cap.isOpened():
    print("video file opening failed")
```

```python
def find_ball_region(path):
    img = cv.imread(path)
    
    width = img.shape[0]
    height = img.shape[1]
    channels = img.shape[2]
    
    print(width, height, channels)
    
    rLen = 0
    for row in range(height):
        for col in range(width):
            
            
            for ch in range(channels):
                img[row][col][ch] += 0
                
    plt.imshow(img)
    plt.axis('off')
```

```python
find_ball_region('./ball_frame.jpg')
```

```python
count = 0
while True:
    ret, frame = cap.read()
    if ret == False:
        break
    cv.imwrite('./ball_frame.jpg', frame)
    blurred = cv.GaussianBlur(frame, (11, 11), 0)
    hsv = cv.cvtColor(blurred, cv.COLOR_BGR2HSV)  
    
    mask = cv.inRange(blurred, redlower, redupper)
    mask = cv.erode(mask, None, iterations = 2)
    mask = cv.dilate(mask, None, iterations = 2)
    count += 1
    
    if count < 10:
        plt.imshow(mask)
        plt.axis('off')
        plt.show()
        break
    
print(count)
```
